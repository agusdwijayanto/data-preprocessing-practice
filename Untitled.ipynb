```python
# 1. Import Libary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import string

nltk.download('punkt')
nltk.download('stopwords')
```

```python
# 2. Import Dataset

df = pd.read_csv('data_penjualan_latihan.csv')

print("5 Data Pertama")
display(df.head())

print("\nInformasi Dataset")
df.info()

print("\nStatistik Deskriptif")
display(df.describe())
```

```python
# 3. Missing Value Analysis

missing = df.isnull().sum()

print("Jumlah Missing Value:")
print(missing)
```

```python
# Visualisasi Missing Value

plt.figure(figsize=(8,4))

missing.plot(kind='bar')

plt.title("Jumlah Missing Value")
plt.xlabel("Kolom")
plt.ylabel("Jumlah")

plt.show()
```

```python
# 4. Data Cleaning

df['harga'] = df['harga'].fillna(df['harga'].mean())

df['jumlah'] = df['jumlah'].fillna(df['jumlah'].mean())

df['rating'] = df['rating'].fillna(df['rating'].mean())

df['deskripsi'] = df['deskripsi'].fillna("tidak ada deskripsi")
```

```python
# Cek Kembali Missing Value

print(df.isnull().sum())
```

```python
# 5. Data Integration

# Membuat Total Harga
df['total_harga'] = df['harga'] * df['jumlah']

# Mengambil Bulan Transaksi
df['tanggal_transaksi'] = pd.to_datetime(df['tanggal_transaksi'])

df['bulan'] = df['tanggal_transaksi'].dt.month

display(df.head())
```

```python
# 6. Data Transformation

scaler = MinMaxScaler()

df['harga_normalisasi'] = scaler.fit_transform(df[['harga']])

df['jumlah_normalisasi'] = scaler.fit_transform(df[['jumlah']])

display(df.head())
```

```python
# Membuat Rating Label

def rating_label(rating):

    if rating <= 2:
        return "Buruk"

    elif rating <= 3:
        return "Cukup"

    else:
        return "Baik"

df['rating_label'] = df['rating'].apply(rating_label)

display(df[['rating', 'rating_label']])
```

```python
# 7. Text Processing

stop_words = set(stopwords.words('indonesian'))
```

```python
# Fungsi Tokenizing dan Filtering

def proses_teks(text):

    tokens = word_tokenize(str(text).lower())

    hasil = []

    for token in tokens:

        if token not in stop_words and token not in string.punctuation:
            hasil.append(token)

    return hasil
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('indonesian'))

def proses_teks(text):

    tokens = str(text).lower().split()

    hasil = []

    for token in tokens:
        if token not in stop_words and token not in string.punctuation:
            hasil.append(token)

    return hasil
```

```python
# Menjalankan Tokenizing

df['token'] = df['deskripsi'].apply(proses_teks)

display(df[['deskripsi', 'token']].head())
```

```python
# 8. Visualisasi
# Grafik Batang Jumlah Produk per Kategori

plt.figure(figsize=(8,4))

df['kategori'].value_counts().plot(kind='bar')

plt.title("Jumlah Produk per Kategori")
plt.xlabel("Kategori")
plt.ylabel("Jumlah Produk")

plt.show()
```

```python
# Grafik Garis Jumlah Transaksi per Bulan

transaksi_bulan = df.groupby('bulan').size()

plt.figure(figsize=(8,4))

plt.plot(
    transaksi_bulan.index,
    transaksi_bulan.values,
    marker='o'
)

plt.title("Jumlah Transaksi per Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Transaksi")

plt.show()
```

```python
# Histogram Harga Produk

plt.figure(figsize=(8,4))

plt.hist(df['harga'], bins=5)

plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")

plt.show()
```
